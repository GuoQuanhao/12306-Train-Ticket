## python 标准库 urllib

### 获取信息

```python
import urllib.request

response = urllib.request.urlopen('http://www.baidu.com/')

print(response.read().decode('utf-8'))
```

### 上传信息

```python
import urllib.request
import urllib.parse

data = bytes(urllib.parse.urlencode({'word':'hello'}), encoding='utf-8')

response = urllib.request.urlopen('http://httpbin.org/post', data=data)
response
html = response.read().decode('utf-8')
print(html)
```

## python 标准库 urllib3

### 获取信息

```python
import urllib3

# 处理与线程的连接以及线程安全
http = urllib3.PoolManager()
response = http.request('GET', 'http://www.baidu.com')

print(response.data.decode())
```

### 上传信息

```python
import urllib3

# 处理与线程的连接以及线程安全
http = urllib3.PoolManager()
response = http.request('POST', 'http://httpbin.org/post', fields={'word':'hello'})

print(response.data.decode())
```

## 第三方库 requests

### 获取特征信息

```python
import requests

response = requests.get('http://www.baidu.com/')

# 防止中文乱码
response.encoding='utf-8'

print('状态码\n', response.status_code)
print('请求地址\n', response.url)
print('头部信息\n', response.headers)
print('Cookie\n', response.cookies)
print('文本源码\n', response.text)
print('字节源码\n', response.content)
```

### 模拟浏览器访问

#### 直接访问被 403 拒绝

```python
import requests

url = 'https://www.whatismyip.com/'

response = requests.get(url)
print(response.status_code)
```

#### 添加 headers

```python
import requests

url = 'https://www.whatismyip.com/'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}

response = requests.get(url, headers=headers)
print(response.content.decode('utf-8'))
```

### 报错信息

设置读取的 timeout 迫使出现网络超时错误

```python
import requests
try:
    response = requests.get('http://www.baidu.com', timeout=0.1)
    print(response.status_code)
except Exception as error:
    print('Timeout:', str(error))
```

设置多个请求错误处理

```python
import requests
from requests.exceptions import ReadTimeout, HTTPError, RequestException
try:
    response = requests.get('https://www.whatismyip.com/', timeout=0.5)
    print(response.status_code)
except ReadTimeout:
    print('Timeout')
except HTTPError:
    print('HttpError')
except RequestException:
    print('RequestError')

try:
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36'}
    response = requests.get('https://www.whatismyip.com/', timeout=1, headers=headers)
    print(response.status_code)
except ReadTimeout:
    print('Timeout')
except HTTPError:
    print('HttpError')
except RequestException:
    print('RequestError')
```

### 设置代理

```python
import requests

proxy = {'https':'119.84.112.139:80', 'http':'101.132.190.101:80'}

response_https = requests.get('https://www.baidu.com', proxies=proxy)
print('response_https:', response_https.status_code)

response_http = requests.get('http://www.baidu.com', proxies=proxy)
print('response_http:', response_http.status_code)
```
